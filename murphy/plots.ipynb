```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

def tempfn(d):
    for key, val in d.items():
        plt.rcParams[key] = val

tempfn({
    'figure.facecolor':'white'
})
```

```python
predictors = pd.read_csv("randomforest/importance_w_desc.csv")
# importance
top50 = predictors.loc[predictors['MeanDecreaseGini'] > 60,:]
fig = plt.figure()
fig.set_size_inches(14, 10)
plt.xticks(rotation=45)
plt.bar(top50['vars'],top50['MeanDecreaseGini'])
plt.title("Variables with Importance > 60")
plt.ylabel("Importance (Gini)")
plt.show()
plt.clf()
```

```python
# % missing
predictors = pd.read_csv("../data/variables.csv")

fig = plt.figure()
fig.set_size_inches(10, 6)
plt.hist(
    predictors['missing_pct'], 
    density= False
)
plt.title("Variable NAs")
plt.xlabel("Missing Percentage")
plt.show()
plt.clf()
```
